```python
products_static_data = pd.read_csv("../results-20190809-152102.csv")
purchases = pd.read_csv("../purchases.csv")
```

```python
purchases.columns =['ts_pur', 'customer_key', 'product_price', 'product_price_net',
       'sku_base', 'sales_promotion_campaign', 'order_total_net',
       'order_total_discount', 'order_total_tax', 'carrier', 'product_type',
       'order_id']
purchases.ts_pur = purchases.ts_pur.apply(pd.Timestamp)
```

```python
products_static_data.dropna(axis = 'columns', how = 'all', inplace = True)

products_static_data = products_static_data[products_static_data.app_id == 'serenata_flowers']
products_static_data = products_static_data[products_static_data.is_addon == False]

products_static_data.drop([
        'units_in_stock', 
        'app_id', 
        'event',
        'is_addon', 
        'sku',
        "fulfillment_code",
        'availability',
        'tax',
        "ts_event",
        "ts_inserted",
        'url',
        'url_image1',
        'url_image2',
        'url_image3',
        'currency',
        'delivery_country',
        'banners',
        'brand',
        'name'        
    ], axis = 'columns', inplace = True)

products_static_data = pd.merge(products_static_data, purchases.groupby('sku_base', as_index=False, ).ts_pur.min(), on = ['sku_base'])
products_static_data = products_static_data.rename(columns={"ts_pur": "first_purchase"}, errors="raise")
```

```python
products_static_data["first_purchase"] = products_static_data.first_purchase.apply(lambda x: pd.Timestamp(x))
products_static_data["week_after_first"] = products_static_data.first_purchase.apply(lambda x: x + pd.offsets.Day(7))
products_static_data["two_weeks_after_first"] = products_static_data.first_purchase.apply(lambda x: x + pd.offsets.Day(14))
products_static_data["four_weeks_after_first"] = products_static_data.first_purchase.apply(lambda x: x + pd.offsets.Day(28))
```

```python
def get_revenue_week(product_name, week = 'week_after_first'):
    date_of_finish = products_static_data[products_static_data.sku_base == product_name][week].unique()[0]
    product_revenue_week_n = purchases[purchases.ts_pur<date_of_finish][purchases.sku_base == product_name].product_price_net.sum() 
    return (product_revenue_week_n)

products_static_data['revenue_after_first_week'] = products_static_data.sku_base.apply(lambda x: get_revenue_week(x))  
products_static_data['revenue_after_second_week'] = products_static_data.sku_base.apply(lambda x: get_revenue_week(x, 'two_weeks_after_first'))  
products_static_data['revenue_after_fours_week'] = products_static_data.sku_base.apply(lambda x: get_revenue_week(x, 'four_weeks_after_first')) 
```
